```python
import numpy as np
import imageio.v2 as imageio
import matplotlib.pyplot as plt
import cv2 as cv
import scipy
from PIL import Image
import random
import time
```

```python
# Read in the images
mountain_base = imageio.imread('images/mountain1.jpg')
mountain_trans = imageio.imread('images/mountain2.jpg')
plt.imshow(mountain_base)
plt.figure()
plt.imshow(mountain_trans)
```

```python
"""
Given matching normalised points from a translated and base image it will calculate the 
homography matrix.

u2_trans: u coordinates of the translated image
v2_trans: v coordinates of the translated image
u_base: u coordinates of the base image
v_base: v coordinates of the base image
returns H: homography matrix
"""
def homography(u2_trans, v2_trans, u_base, v_base):
    
    # A matrix with size (N * 2) x 9
    a_mat = np.zeros((len(u_base) * 2, 9))
    uv_base = np.dstack((u_base, v_base))[0]
    uv_trans = np.dstack((u2_trans, v2_trans))[0]
    # iterate through corresponding points
    for i, x_i in enumerate(uv_base):
        # each set of corresponding points can generate two equations
        # basic_block is the xyw points of the base coordinates
        basic_block = np.array([x_i[0], x_i[1], 1])
        block1 = basic_block # w = 1
        block2 = basic_block * uv_trans[i][1]
        block3 = -basic_block * uv_trans[i][0]
        zero_block = np.zeros(3)
        part = np.vstack((
            np.hstack((zero_block, -block1, block2)),
            np.hstack((block1, zero_block, block3)),
        ))
        a_mat[2 * i: 2 * (i + 1)] = part

    # SVD to matrix A
    v = np.linalg.svd(a_mat)[2]
    # h should be the last column of matrix V and rearrange it into matrix P
    H = v[-1].reshape(3,3)
    # normalize H 
    H = H / np.linalg.norm(H)
    return H
```

```python
"""
Produce the normalisation matrix given the height and width of the image
"""
def normalisation_mat(w, h):
    return np.linalg.inv(
        np.array([
            [w + h, 0, w / 2],
            [0, w + h, h / 2],
            [0, 0, 1]
        ])
    )

"""
Produce a unnormalised homography matrix given corresponding homogenous coordinates from the base and translated image
img_base: base image numpy array
img_trans: translated image numpy array
base_points: homogenous coordinates from the base image
trans_points: homogenous coordinates from the translated image
returns H: unnormalised homography matrix
"""
def homography_w_normalisation(img_base, img_trans, base_points, trans_points):
    # Get Normalisation matrix
    base_norm = normalisation_mat(img_base.shape[1], img_base.shape[0])
    trans_norm = normalisation_mat(img_trans.shape[1],img_trans.shape[0])
    
    # Normalise points
    uv_base_norm = (base_norm @ base_points.T).T
    uv_trans_norm = (trans_norm @ trans_points.T).T

    # Calculate homography from normalised points
    H_hat = homography(uv_trans_norm[:,0], uv_trans_norm[:,1], uv_base_norm[:,0], uv_base_norm[:,1])
    
    # Denormalise solution
    H =  np.linalg.inv(trans_norm) @ H_hat @ base_norm
    return H
    
```

```python
# Convert images to grayscale
mountain1_gray = np.uint8(np.dot(mountain_base[:,:,:3], [0.299, 0.587, 0.114]))
mountain2_gray = np.uint8(np.dot(mountain_trans[:,:,:3], [0.299, 0.587, 0.114]))

# SIFT to find matching points in both images
# https://docs.opencv.org/4.x/da/df5/tutorial_py_sift_intro.html
sift = cv.SIFT_create()
kp1,des1 = sift.detectAndCompute(mountain1_gray,None)
kp2,des2 = sift.detectAndCompute(mountain2_gray,None)

# https://docs.opencv.org/4.x/dc/dc3/tutorial_py_matcher.html
# create BFMatcher object
bf = cv.BFMatcher()
# Match descriptors.
matches = bf.knnMatch(des1,des2, k=2)

# Apply ratio test
good = []
for m,n in matches:
    if m.distance < 0.75 * n.distance:
        good.append([m])

# cv.drawMatchesKnn expects list of lists as matches.
img3 = cv.drawMatchesKnn(mountain_base,kp1,mountain_trans,kp2,good,None,flags=cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
plt.figure()
plt.title("SIFT Feature Descriptor")
plt.imshow(img3)
plt.show()

# https://stackoverflow.com/questions/30716610/how-to-get-pixel-coordinates-from-feature-matching-in-opencv-python
# Get the coordinates from the matches
list_kp1 = [kp1[mat[0].queryIdx].pt for mat in good] 
list_kp2 = [kp2[mat[0].trainIdx].pt for mat in good]

# Sample half the points using seed 42
random.seed(42)
N = len(list_kp1)
points = random.sample(range(0, len(list_kp1)), int(np.ceil(len(list_kp1)/2)))
# Get the 
base_points = np.array([list_kp1[i] for i in points])
transformed_points = np.array([list_kp2[i] for i in points])

# Store npy file
np.save("base.npy", base_points)
np.save("trans.npy", transformed_points)
```

```python
"""
Create homography matrix using the RANSAC algorithm
N: number of iterations
distance_thresh: the threshold for distance between translated image points and the calculated
projection points given the H matrix
inlier_thresh: Threshold for what percent of inliers we want to stop at (between 0-1)
img_base: base image array
img_trans: translated image array

returns H: homography matrix
"""
def homography_w_normalisation_ransac(N, distance_thresh, inlier_thresh, img_base, img_trans):
    base = np.load("base.npy")
    trans = np.load("trans.npy")
    l = len(base)
    # Convert all coordinates to homogenous coordinates
    base_homo = np.hstack((base, np.ones((base.shape[0], 1))))
    trans_homo = np.hstack((trans, np.ones((trans.shape[0], 1))))
    num_best_inliers = 0
    
    for i in range(N):
        # Sample 4 randomly
        idx = random.sample(range(l), 4)
        base_points = base_homo[idx]
        trans_points = trans_homo[idx]
        
        # get homography matrix from sample
        H = homography_w_normalisation(img_base, img_trans, base_points, trans_points)
        
        # Matrix that holds the projected uv coordinates
        projected_uv = np.zeros((l,2))
        # Compute how many inliers
        for j in range(l):
            p = H @ base_homo[j]
            # Divide x and y by w
            projected_uv[j][0] = p[0]/p[2]
            projected_uv[j][1] = p[1]/p[2]
        # Compute distance between generated point and ground truth
        inlier = np.linalg.norm((trans - projected_uv), axis = 1)
        
        # npwhere for how many inliers d<t
        inlier[inlier > distance_thresh] = 0
        inliers = np.argwhere(inlier)
        
        # If larger than inlier threshold  return the current H
        if len(inliers)/l > inlier_thresh:
            return H
        # Replace current best if better
        if len(inliers) > num_best_inliers:
            num_best_inliers = len(inliers)
            best_H = H.copy()
    return best_H
        
```

```python
# Get homography matrix
H = homography_w_normalisation_ransac(3000, 3, 0.95, mountain_base, mountain_trans)

# Warp the first image given the homography 
dst = cv.warpPerspective(mountain_base,H,((mountain_base.shape[1] + mountain_trans.shape[1]), mountain_trans.shape[0]))
# Stitch second image onto warped base image
dst[0:mountain_trans.shape[0], 0:mountain_trans.shape[1]] = mountain_trans
plt.figure()
plt.title("My Homography")
plt.imshow(dst)
plt.show()

# Check my output with CV output
base = np.load("base.npy")
trans = np.load("trans.npy")
# https://docs.opencv.org/3.4/d9/d0c/group__calib3d.html#ga4abc2ece9fab9398f2e560d53c8c9780
M, _ = cv.findHomography(base, trans, cv.RANSAC, 3, maxIters = 3000)
dst = cv.warpPerspective(mountain_base,M,((mountain_base.shape[1] + mountain_trans.shape[1]), mountain_trans.shape[0]))
dst[0:mountain_trans.shape[0], 0:mountain_trans.shape[1]] = mountain_trans
plt.figure()
plt.title("CV Homography")
plt.imshow(dst)
plt.show()
```
